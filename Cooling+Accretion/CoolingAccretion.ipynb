```python
from likelihood import *
from detection import *
from Profile_and_Contours import *
import matplotlib.pyplot as plt
import numpy as np
import time
from iminuit import Minuit
from scipy.stats import chi2   
import os.path
import pandas as pd 
import itertools
plt.style.use('style.mplstyle')
```

```python
#Parameters allowed space
#Cooling
Tci, Tcf = 1, 10
tau_ci, tau_cf = 1, 40
Rci, Rcf = 1, 100
tau_i, tau_f = 1, 1.4

#Off-set times
toff_K_i,toff_K_f =0,6
toff_I_i,toff_I_f =0,6
toff_B_i,toff_B_f =0,6

#Accretion
Tai, Taf = 0.1, 7
tau_ai, tau_af = 0.3, 3.5
Mai, Maf = 0, 0.6

#nubar_e survival probability 
Pee_i,Pee_f=0,1
```

```python
#Cooling+Accreation
#NH
Likelihood_c_a_NH.errordef = Minuit.LEAST_SQUARES
x0_c_a_NH=np.array([4.81,4.16,12.3,1.2,2.0,0.57,0.6,0.0,0.47,0.0])
m_c_a_NH = Minuit(Likelihood_c_a_NH, x0_c_a_NH)
m_c_a_NH.limits = [(Tci, Tcf), (tau_ci, tau_cf), (Rci, Rcf),(tau_i, tau_f), (Tai, Taf), (tau_ai, tau_af), (Mai, Maf),(toff_K_i,toff_K_f),(toff_I_i,toff_I_f),(toff_B_i,toff_B_f)]
m_c_a_NH.fixed["x3"] = True
#IH
Likelihood_c_a_IH.errordef = Minuit.LEAST_SQUARES
x0_c_a_IH=np.array([3.98,4.62,14.5,1.2,3.12,0.67,0.6,0.0,0.0,0.0])
m_c_a_IH = Minuit(Likelihood_c_a_IH, x0_c_a_IH)
m_c_a_IH.limits = [(Tci, Tcf), (tau_ci, tau_cf), (Rci, Rcf),(tau_i, tau_f), (Tai, Taf), (tau_ai, tau_af), (Mai, Maf),(toff_K_i,toff_K_f),(toff_I_i,toff_I_f),(toff_B_i,toff_B_f)]
m_c_a_IH.fixed["x3"] = True
#Pee
Likelihood_c_a_Pee.errordef = Minuit.LEAST_SQUARES
x0_c_a_Pee=np.array([5.37 ,4.13,11.3,1.2,1.91,0.58,0.6,0.0,0.47,0.0,1])
m_c_a_Pee = Minuit(Likelihood_c_a_Pee, x0_c_a_Pee)
m_c_a_Pee.limits = [(Tci, Tcf), (tau_ci, tau_cf), (Rci, Rcf),(tau_i, tau_f), (Tai, Taf), (tau_ai, tau_af), (Mai, Maf),(toff_K_i,toff_K_f),(toff_I_i,toff_I_f),(toff_B_i,toff_B_f),(Pee_i, Pee_f)]
m_c_a_Pee.fixed["x3"] = True
```

## 2.2. Best Fit Minimization <a class="anchor" id="2.2-bullet"></a>

```python
#Cooling+Accreation
#NH
start_time = time.time()
m_c_a_NH.migrad()
m_c_a_NH_BF=np.array(m_c_a_NH.values)
m_c_a_NH_BF_like=np.array(m_c_a_NH.fval)
np.save("Output/BF_NH_TD_coo_acc.npy", np.array(m_c_a_NH_BF,dtype=object))
np.save("Output/BF_like_NH_TD_coo_acc.npy", np.array(m_c_a_NH_BF_like,dtype=object))
# m_c_a_NH.minos()
print("Minuit NH= %s minutes" % ((time.time() - start_time)/60))
# m_c_a_NH.fixed["x6"] = True  #M_a
m_c_a_NH.fixed["x7"] = True  #toff_K
m_c_a_NH.fixed["x8"] = True  #toff_I
m_c_a_NH.fixed["x9"] = True  #toff_B
print(m_c_a_NH.fval)
print(m_c_a_NH.params)

#IH
start_time = time.time()
m_c_a_IH.migrad()
m_c_a_IH_BF=np.array(m_c_a_IH.values)
m_c_a_IH_BF_like=np.array(m_c_a_IH.fval)
np.save("Output/BF_IH_TD_coo_acc.npy", np.array(m_c_a_IH_BF,dtype=object))
np.save("Output/BF_like_IH_TD_coo_acc.npy", np.array(m_c_a_IH_BF_like,dtype=object))
# m_c_a_IH.minos()
print("Minuit IH= %s minutes" % ((time.time() - start_time)/60))
# m_c_a_IH.fixed["x6"] = True  #M_a
m_c_a_IH.fixed["x7"] = True  #toff_K
m_c_a_IH.fixed["x8"] = True  #toff_I
m_c_a_IH.fixed["x9"] = True  #toff_B
print(m_c_a_IH.fval)
print(m_c_a_IH.params)
```

```python
params=["Tc","tau_c","Rc",'Ta','tau_a',"Ma"]
params_index=[0,1,2,4,5,6]
params_lim=[(Tci, Tcf), (tau_ci, 25), (Rci, 70), 
            (1, 5.5), (tau_ai, tau_af), (Mai, Maf)]
params_label=[r"$T_c$ [MeV]",r"$\tau_c$ [s]",r"$R_c$ [km]",r"$T_a$ [MeV]",r"$\tau_a$ [s]",r"$M_a$ [$M_\odot$]"]
```

```python
x_c_a_NH_prof,params_c_a_NH_prof,L_c_a_NH_prof=[],[],[]
x_c_a_IH_prof,params_c_a_IH_prof,L_c_a_IH_prof=[],[],[]
x_c_a_Pee_prof,params_c_a_Pee_prof,L_c_a_Pee_prof=[],[],[]

n_iterate=100
n_points=30

for i in range(len(params)):
    start_time = time.time()
    #NH
    m_c_a_NH.values=m_c_a_NH_BF
    x, L, params_aux = param_profile(m_c_a_NH,"x%d"%params_index[i],params_lim[i][0],params_lim[i][1],size=n_points,iterate=n_iterate,subtract_min=True)
    x_c_a_NH_prof.append(x)
    L_c_a_NH_prof.append(L)
    params_c_a_NH_prof.append(params_aux)

    #IH
    m_c_a_IH.values=m_c_a_IH_BF
    x, L, params_aux = param_profile(m_c_a_IH,"x%d"%params_index[i],params_lim[i][0],params_lim[i][1],size=n_points,iterate=n_iterate,subtract_min=True)
    x_c_a_IH_prof.append(x)
    L_c_a_IH_prof.append(L)
    params_c_a_IH_prof.append(params_aux)

    #Pee
    m_c_a_Pee.values=m_c_a_Pee_BF
    x, L, params_aux = param_profile_mixed(m_c_a_Pee,m_c_a_NH,m_c_a_IH,"x%d"%params_index[i],params_lim[i][0],params_lim[i][1],size=n_points,iterate=n_iterate,subtract_min=True)
    x_c_a_Pee_prof.append(x)
    L_c_a_Pee_prof.append(L)
    params_c_a_Pee_prof.append(params_aux)
        
    print("Minuit ",params[i]," : ",((time.time() - start_time)/60)," minutes")

#Saving Data
#NH
np.save("Output/x_NH_TD_coo_acc_prof.npy", np.array(x_c_a_NH_prof,dtype=object))
np.save("Output/Like_NH_TD_coo_acc_prof.npy", np.array(L_c_a_NH_prof,dtype=object))
#IH
np.save("Output/x_IH_TD_coo_acc_prof.npy", np.array(x_c_a_IH_prof,dtype=object))
np.save("Output/Like_IH_TD_coo_acc_prof.npy", np.array(L_c_a_IH_prof,dtype=object))
#Pee
np.save("Output/x_Pee_TD_coo_acc_prof.npy", np.array(x_c_a_Pee_prof,dtype=object))
np.save("Output/Like_Pee_TD_coo_acc_prof.npy", np.array(L_c_a_Pee_prof,dtype=object))
```
