```python
from likelihood import *
from detection import *
from Profile_and_Contours import *
import matplotlib.pyplot as plt
import numpy as np
import time
from iminuit import Minuit
from scipy.stats import chi2   
import os.path
import pandas as pd 
import itertools
plt.style.use('style.mplstyle')

################################################################
#Parameters allowed space
#Cooling
Tci, Tcf = 1, 10
tau_ci, tau_cf = 1, 40
Rci, Rcf = 1, 100
tau_i, tau_f = 1, 1.4

#Off-set times
toff_K_i,toff_K_f =0,6
toff_I_i,toff_I_f =0,6
toff_B_i,toff_B_f =0,6

#nubar_e survival probability 
Pee_i,Pee_f=0,1


################################################################
#Only Cooling
#NH
Likelihood_c_NH.errordef = Minuit.LEAST_SQUARES
x0_c_NH=[3.27,5.1,62,1.2,0,0,0]
m_c_NH = Minuit(Likelihood_c_NH, x0_c_NH)
m_c_NH.limits = [(Tci, Tcf), (tau_ci, tau_cf), (Rci, Rcf),(tau_i, tau_f),(toff_K_i,toff_K_f),(toff_I_i,toff_I_f),(toff_B_i,toff_B_f)]
m_c_NH.fixed["x3"] = True

#IH
Likelihood_c_IH.errordef = Minuit.LEAST_SQUARES
x0_c_IH=[3.65,5.1,31,1.2,0,0,0]
m_c_IH = Minuit(Likelihood_c_IH, x0_c_IH)
m_c_IH.limits = [(Tci, Tcf), (tau_ci, tau_cf), (Rci, Rcf),(tau_i, tau_f),(toff_K_i,toff_K_f),(toff_I_i,toff_I_f),(toff_B_i,toff_B_f)]
m_c_IH.fixed["x3"] = True

#Pee
Likelihood_c_Pee.errordef = Minuit.LEAST_SQUARES
x0_c_Pee=[3.6,40,31,1.2,0,0,0,0.5]
m_c_Pee = Minuit(Likelihood_c_Pee, x0_c_Pee)
m_c_Pee.limits = [(Tci, Tcf), (tau_ci, tau_cf), (Rci, Rcf),(tau_i, tau_f),(toff_K_i,toff_K_f),(toff_I_i,toff_I_f),(toff_B_i,toff_B_f),(Pee_i, Pee_f)]
m_c_Pee.fixed["x3"] = True

################################################################
#NH
start_time = time.time()
m_c_NH.migrad()
print("Minuit NH= %s minutes" % ((time.time() - start_time)/60)) 
m_c_NH_BF=np.array(m_c_NH.values)
m_c_NH_BF_like=np.array(m_c_NH.fval)
np.save("Output/BF_NH_TD_coo.npy", np.array(m_c_NH_BF,dtype=object))
np.save("Output/BF_like_NH_TD_coo.npy", np.array(m_c_NH_BF_like,dtype=object))
m_c_NH.fixed["x4"] = True
m_c_NH.fixed["x5"] = True
m_c_NH.fixed["x6"] = True
print(m_c_NH)

#IH
start_time = time.time()
m_c_IH.migrad()
print("Minuit IH= %s minutes" % ((time.time() - start_time)/60))
m_c_IH_BF=np.array(m_c_IH.values)
m_c_IH_BF_like=np.array(m_c_IH.fval)
np.save("Output/BF_IH_TD_coo.npy", np.array(m_c_IH_BF,dtype=object))
np.save("Output/BF_like_IH_TD_coo.npy", np.array(m_c_IH_BF_like,dtype=object))
m_c_IH.fixed["x4"] = True
m_c_IH.fixed["x5"] = True
m_c_IH.fixed["x6"] = True
print(m_c_IH)

#Pee
start_time = time.time()
m_c_Pee.migrad()
print("Minuit Pee= %s minutes" % ((time.time() - start_time)/60))
m_c_Pee_BF=np.array(m_c_Pee.values)
m_c_Pee_BF_like=np.array(m_c_Pee.fval)
np.save("Output/BF_Pee_TD_coo.npy", np.array(m_c_Pee_BF,dtype=object))
np.save("Output/BF_like_Pee_TD_coo.npy", np.array(m_c_Pee_BF_like,dtype=object))
m_c_Pee.fixed["x4"] = True
m_c_Pee.fixed["x5"] = True
m_c_Pee.fixed["x6"] = True
print(m_c_Pee)
```

```python
params=["Tc","tau_c","Rc"]
params_index=[0,1,2]
params_lim=[(2,8), (tau_ci, 15), (Rci, Rcf)]
params_label=[r"$T_c$ [MeV]",r"$\tau_c$ [s]",r"$R_c$ [km]"]

x_NH_prof,L_NH_prof,params_NH_prof=[],[],[]
x_IH_prof,L_IH_prof,params_IH_prof=[],[],[]
x_Pee_prof,L_Pee_prof,params_Pee_prof=[],[],[]
```

```python
n_iterate=20
n_points=30

for i in params_index:
    start_time = time.time()
    #NH
    m_c_NH.values=m_c_NH_BF
    x, L, params_aux = param_profile(m_c_NH,"x%d"%i,params_lim[i][0],params_lim[i][1],size=n_points,iterate=n_iterate,subtract_min=True)
    x_NH_prof.append(x)
    L_NH_prof.append(L)
    params_NH_prof.append(params_aux)

    #IH
    m_c_IH.values=m_c_IH_BF
    x, L, params_aux = param_profile(m_c_IH,"x%d"%i,params_lim[i][0],params_lim[i][1],size=n_points,iterate=n_iterate,subtract_min=True)
    x_IH_prof.append(x)
    L_IH_prof.append(L)
    params_IH_prof.append(params_aux)

    #Pee
    m_c_Pee.values=m_c_Pee_BF
    x, L, params_aux = param_profile(m_c_Pee,"x%d"%i,params_lim[i][0],params_lim[i][1],size=n_points,iterate=n_iterate,subtract_min=True)
    x_Pee_prof.append(x)
    L_Pee_prof.append(L)
    params_Pee_prof.append(params_aux)
        
    print("Minuit ",params[i]," : ",((time.time() - start_time)/60)," minutes")
```

```python
#Saving Data
#NH
np.save("Output/x_NH_TD_coo_prof.npy", np.array(x_NH_prof,dtype=object))
np.save("Output/Like_NH_TD_coo_prof.npy", np.array(L_NH_prof,dtype=object))
#IH
np.save("Output/x_IH_TD_coo_prof.npy", np.array(x_IH_prof,dtype=object))
np.save("Output/Like_IH_TD_coo_prof.npy", np.array(L_IH_prof,dtype=object))
#Pee
np.save("Output/x_Pee_TD_coo_prof.npy", np.array(x_Pee_prof,dtype=object))
np.save("Output/Like_Pee_TD_coo_prof.npy", np.array(L_Pee_prof,dtype=object))
```
